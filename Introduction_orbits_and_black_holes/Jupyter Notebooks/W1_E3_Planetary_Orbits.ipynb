```python
import numpy as np
import matplotlib.pyplot as plt

def get_radius(theta_list, e, r0):
    return r0 / (1 + e * np.cos(theta_list))

theta = np.linspace(0, 2 * np.pi, 500)

# 1. Circle: e = 0
x_c = get_radius(theta, 0, 1) * np.cos(theta)
y_c = get_radius(theta, 0, 1) * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x_c, y_c, 'b', lw=2)
plt.scatter([0], [0], color='yellow', s=100, edgecolors='orange', label='Sun')
plt.gca().set_aspect('equal')
plt.title('1. Circle ($e=0$)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='upper right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_radius(theta_list, e, r0):
    return r0 / (1 + e * np.cos(theta_list))

theta = np.linspace(0, 2 * np.pi, 500)

# 2. Ellipse: e = 1/2
x_e = get_radius(theta, 1/2, 1) * np.cos(theta)
y_e = get_radius(theta, 1/2, 1) * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x_e, y_e, 'g', lw=2)
plt.scatter([0], [0], color='yellow', s=100, edgecolors='orange', label='Sun')
plt.gca().set_aspect('equal')
plt.title('2. Ellipse ($0 < e < 1$)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_radius(theta_list, e, r0):
    return r0 / (1 + e * np.cos(theta_list))

# 3. Hyperbola: e = 2
e_hyp = 2
limit_phi_hyp = np.arccos(-1/e_hyp)
phi_hyp = np.linspace(-limit_phi_hyp + 0.01, limit_phi_hyp - 0.01, 1000)

x_h = get_radius(phi_hyp, e_hyp, 1) * np.cos(phi_hyp)
y_h = get_radius(phi_hyp, e_hyp, 1) * np.sin(phi_hyp)

plt.figure(figsize=(6, 6))
plt.plot(x_h, y_h, 'k', lw=2)
plt.scatter([0], [0], color='yellow', s=100, edgecolors='orange', label='Sun')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal')
plt.title('3. Hyperbola ($e > 1$)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_radius(theta_list, e, r0):
    return r0 / (1 + e * np.cos(theta_list))

# 4. Parabola: e = 1
limit_phi_par = np.arccos(-1)
phi_par = np.linspace(-limit_phi_par + 0.1, limit_phi_par - 0.1, 1000)

x_p = get_radius(phi_par, 1, 1) * np.cos(phi_par)
y_p = get_radius(phi_par, 1, 1) * np.sin(phi_par)

plt.figure(figsize=(6, 6))
plt.plot(x_p, y_p, 'darkorange', lw=2)
plt.scatter([0], [0], color='yellow', s=100, edgecolors='orange', label='Sun')
plt.ylim(-3, 3)
plt.xlim(-2, 2)
plt.title('4. Parabola ($e = 1$)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```
