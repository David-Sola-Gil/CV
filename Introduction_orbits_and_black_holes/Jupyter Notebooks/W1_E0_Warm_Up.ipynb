```python
# Import Python libraries
import numpy as np #Mathematical functions and vectors
import matplotlib.pyplot as plt #Plots and graphics

# Compute the vectors/lists required for the plot
x = np.linspace(-5, 5, 100)
y = x**2

# Ask Python to draw the plot, choosing colour and legend
plt.plot(x,y, color='red')
plt.legend(['$y=x^2$'], loc="upper right")
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.grid() # Add a grid
plt.show()
```
