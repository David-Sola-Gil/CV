```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
#mpl.rcParams['figure.dpi'] = 600

# Constants for a nice plot
m = 1.0     # Mass
k = 10.0    # Constant k = G * M
l = 2.0     # Angular momentum per unit mass

# Function to compute the effective potential
def effective_potential(r):
    attractive = -m * k / r
    repulsive = (m * l**2) / (2 * r**2)
    return attractive + repulsive

# Shorten the range of r to avoid division by 0
r = np.linspace(0.01, 6.0, 500)
v_eff = effective_potential(r)

# Compute the minimum to show it on the plot
r_min = l**2 / k
v_min = effective_potential(r_min)

# Graphics
plt.figure(figsize=(10, 6))
plt.plot(r, v_eff, label=r'$V_{eff}(r)$', color='purple', lw=2)

# Add the two terms in the potential to the plot, for comparison
plt.plot(r, -m * k / r, '--', label='Atractive ($-10/r$)', alpha=0.5, color='red')
plt.plot(r, (m * l**2) / (2 * r**2), '--', label='Repulsive ($2/r^2$)', alpha=0.5,color='blue')

# Stable orbit point
plt.plot(r_min, v_min, 'ro')

# Display the final graph, adding horizontal lines to discuss different types of orbits
plt.axhline(effective_potential(r_min), color='orange', linewidth=1, label='E0')
plt.axhline(-8, color='orange', linewidth=1, label='E1')
plt.axhline(5, color='orange', linewidth=1, label='E2')
plt.axhline(0, color='orange', linewidth=1, label='E3')
plt.ylim(v_min - 5, 20)
plt.xlim(0, 6)
plt.xlabel('r (m)')
plt.ylabel('Energy (J)')
plt.title('Effective potential using Newtonian gravity.')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)

plt.show()
```
