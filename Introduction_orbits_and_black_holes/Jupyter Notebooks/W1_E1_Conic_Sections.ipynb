```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Circle: x^2 + y^2 = r^2
r = 5
theta = np.linspace(0, 2 * np.pi, 500)
x_c = r * np.cos(theta)
y_c = r * np.sin(theta)

plt.figure(figsize=(7, 6))
plt.plot(x_c, y_c, 'b', lw=2, label=f'Circle: $x^2 + y^2 = {r}^2$')
plt.plot([0, r], [0, 0], 'r--', marker='o', label=f'Radius $r = {r}$')
plt.gca().set_aspect('equal')
plt.title('1. Circle')
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2. Ellipse: (x^2 / a^2) + (y^2 / b^2) = 1
a_e, b_e = 6, 5
theta = np.linspace(0, 2 * np.pi, 500)
x_e = a_e * np.cos(theta)
y_e = b_e * np.sin(theta)
d_foco_e = np.sqrt(a_e**2 - b_e**2)

plt.figure(figsize=(7, 6))
plt.plot(x_e, y_e, 'g', lw=2, label=rf'Ellipse: $\frac{{x^2}}{{{a_e}^2}} + \frac{{y^2}}{{{b_e}^2}} = 1$')
plt.plot([0, a_e], [0, 0], 'r--', label=f'Semi-major axis $a = {a_e}$')
plt.plot([0, 0], [0, b_e], 'm--', label=f'Semi-minor axis $b = {b_e}$')
plt.scatter([-d_foco_e, d_foco_e], [0, 0], color='yellow', zorder=5, edgecolors='black', label='Foci')
plt.gca().set_aspect('equal')
plt.title('2. Ellipse')
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 3. Hyperbola: (x^2 / a^2) - (y^2 / b^2) = 1
a_h, b_h = 2, 1.5
t = np.linspace(-2, 2, 100)
x_hr = a_h * np.cosh(t)
y_hr = b_h * np.sinh(t)
x_hl = -a_h * np.cosh(t)
y_hl = b_h * np.sinh(t)
d_foco_h = np.sqrt(a_h**2 + b_h**2)
x_dir_h = a_h**2 / d_foco_h

plt.figure(figsize=(7, 6))
plt.plot(x_hr, y_hr, 'k', lw=2, label=rf'Hyperbola: $\frac{{x^2}}{{{a_h}^2}} - \frac{{y^2}}{{{b_h}^2}} = 1$')
plt.plot(x_hl, y_hl, 'k', lw=2)
plt.plot([0, a_h], [0, 0], 'r--', label=f'Distance to vertex $a = {a_h}$')
plt.scatter([-d_foco_h, d_foco_h], [0, 0], color='yellow', zorder=5, edgecolors='black', label='Foci')

# Asíntotas y directrices
x_range = np.linspace(-8, 8, 100)
plt.plot(x_range, (b_h/a_h)*x_range, 'y:', alpha=0.6, label=r'Asymptotes $y = \pm \frac{b}{a}x$')
plt.plot(x_range, -(b_h/a_h)*x_range, 'y:', alpha=0.6)
plt.vlines([-x_dir_h, x_dir_h], ymin=-6, ymax=6, color='purple', linestyle='--', label=rf'Directrices $x = \pm {x_dir_h:.2f}$')

plt.xlim(-8, 8)
plt.ylim(-6, 6)
plt.gca().set_aspect('equal')
plt.title('3. Hyperbola')
plt.legend(loc='upper right', fontsize='small')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 4. Parabola: x^2 = 4py -> y = x^2 / (4p)
p = 1.5
x_p = np.linspace(-8, 8, 100)
y_p = (x_p**2) / (4 * p)

plt.figure(figsize=(7, 6))
plt.plot(x_p, y_p, 'darkorange', lw=2, label=rf'Parabola: $x^2 = 4({p})y$')
plt.scatter([0], [p], color='yellow', zorder=5, edgecolors='black', label=f'Focus $(0, {p})$')
plt.axhline(-p, color='purple', linestyle='--', label=f'Directrix $y = -{p}$')
plt.ylim(-3, 8)
plt.title('4. Parabola')
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
